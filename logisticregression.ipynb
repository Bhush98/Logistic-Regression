```python
# ----------------------------------------------------------------------------------------
# Predicting Breast cancer using Logistic regression whether cancer is benign or melignant
# ----------------------------------------------------------------------------------------

# Logistic regression is a classification regression algorithm used for both classification as
# well as regression the difference in linear regression and Logistic regression is Linear
# regression predicts a value according to the given data and logistic regression classifies a 
# given instance either of a two classification (Binary) we can also do multiclass classification 
# using Logistic regression both Linear as well as Logistic regression uses Linear equations to solve
# problem whether regression or classification
# logitic regression forms a 'S' curve on a graph if the outcome value is less than 0.5 or towards negative
# the outcome is zero else the outcome if it is towards +ve or towards 1 it is 1

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Downloads/data.csv') #reading our data in the form of csv file
```

```python
df.head() #printing our first five instances
```

```python
print(df) #printing our entire dataset
```

```python
X = df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']]
#getting our X data
```

```python
print(X) #printing our X data
```

```python
 y = df['diagnosis'] # getting our y data that is the lassification whether the cancer is benign or melignant
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y) #splitting our entire data into training and testing
```

```python
from sklearn.linear_model import LogisticRegression #importing our model fro sklearn
```

```python
logreg = LogisticRegression(C=5).fit(x_train,y_train) #fitting our training data into the classifier
```

```python
accuracy_score = logreg.score(x_test,y_test) #getting the accuracy score
```

```python
print(accuracy_score) #printing accuracy score
```

```python
# accuracy score defines how well our algorithm can predict the outcome of a new instance
# In this case given a set of new measurements the Algorithm predicts whether the cancer is 
# harmful or harmless
# melignant or Benign
```
